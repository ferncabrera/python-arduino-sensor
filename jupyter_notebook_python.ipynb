```python
import pyfirmata as pf
import numpy as np
import matplotlib as plt
import time
```

```python
port = 'COM5'
board = pf.Arduino(port)
```

```python
board.analog[0].mode = pf.INPUT
```

```python
it = pf.util.Iterator(board)
it.start()
```

```python
board.analog[0].enable_reporting()
```

```python
i = 0
a = 0.00385
vb = 1.559
ro = 200
to = 0
temperatures = []
while i<720:
    pin_0 = board.analog[0].read()
    if pin_0 == None:
        time.sleep(10)
        print(".1.")
        i += 1
        
    else:
        time.sleep(10)
        volts = pin_0 * 5.
        print (volts)
        current =(vb-volts)/ro
        print (current)
        rtd_rest = volts/current
        print(rtd_rest)
        temp = (((rtd_rest/ro)-1)/a)
        print(temp-14)
        temperatures.append(temp)
        if (i==90 or i==180 or i==270 or i==360 or i==450 or i==540 or i==630 or i==720):
            vb = vb-0.005
        i += 1
        
```
